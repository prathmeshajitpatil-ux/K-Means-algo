```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Step 1: Create sample dataset
X = np.array([
    [1,2],[1.5,1.8],[5,8],[8,8],[1,0.6],
    [9,11],[8,2],[10,2],[9,3]
])

# Step 2: Apply K-Means
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# Step 3: Get cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Step 4: Plot data points
plt.scatter(X[:,0], X[:,1], c=labels, cmap='rainbow', s=100)

# Step 5: Plot centroids
plt.scatter(centroids[:,0], centroids[:,1], color='black', marker='X', s=200)

# Step 6: Graph title
plt.title("K-Means Clustering")

# Step 7: Show graph
plt.show()
```
